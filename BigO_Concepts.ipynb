# 1️⃣ **O(1) - Constant Time**

Steps:
- Retrieve the first element using lst[0].


O(1) → The operation is constant regardless of the list size.


```python
def get_the_first_element(lst):
  return lst[0]

numbers = [10,20,30,40,50]

result = get_the_first_element(numbers)
print(result)
```

# **2️⃣ O(log n) - Logarithmic Time**


Steps:
- Find the middle index.
- Compare the middle element with target.
- If target is smaller, search in the left half.
- If target is larger, search in the right half.
- Repeat steps 1-4 until target is found or the range is empty.


Time Complexity:
- O(log n) → Each step halves the search space.

The power Of Log n -- you can find any number at 31 steps.

```python
def binary_search(arr,target):
  left, right = 0, len(arr)-1
  while left <= right:
    mid = (left + right) //2
    if arr[mid] == target:
      return arr[mid]
    elif arr[mid] < target:
      left =  mid + 1
    else:
      right = mid - 1
  return -1


numbers = [1, 3, 5, 7, 9, 11, 13, 15]
print(binary_search(numbers, 3))
```

# **3️⃣ O(n) - Linear Time**

Steps:
- Loop through each element.
- Print the element.


O(n) → The loop runs once for each item.



```python
def print_elements(arr):
  for element in arr:
    print(element)

App = ['Facebook','Instapay','Whatapp']

show_App = print_elements(App)
print(show_App)
```

# **4️⃣ O(n log n) - Linearithmic Time**

Sorting Algorithm

Steps:
- Divide the list into halves.
- Recursively sort each half.
- Merge sorted halves.

Time Complexity:
- O(n log n) → Splitting (log n), merging (n).

```python
def merge_sort(arr):
    if len(arr) <= 1 :
        return arr

    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    sorted_arr = []
    l = r =0


    while len(left_half) > l  and len(right_half) > r:

        if left_half[l] < right_half[r]:
            sorted_arr.append(left_half[l])
            l +=1
        else:
            sorted_arr.append(right_half[r])
            r +=1

    sorted_arr.extend(right_half[r:])
    sorted_arr.extend(left_half[l:])


    return sorted_arr

numbers = [10, 3, 15, 7, 8, 23, 74, 18]
print(merge_sort(numbers))  # O(n log n)
```

# **5️⃣ O(n²) - Quadratic Time**

Steps:
- Loop through the list.
- Compare adjacent elements.
- Swap if out of order.
- Repeat until sorted.


Time Complexity:
- O(n²) → Each element compares with every other.

```python
def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
    for j in range(0, n-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
  return arr


numbers = [5, 1, 4, 2, 8]
print(bubble_sort(numbers))  # O(n²)
```
